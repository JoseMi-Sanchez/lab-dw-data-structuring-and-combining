```python
import pandas as pd

df_cleaned = pd.read_csv('cleaned_cars_policies.csv')

# Verificar que se ha cargado correctamente
print("Dataset limpio del ejercicio anterior:")
df1 = df_cleaned
df1
```

```python
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

# Cargar los nuevos datos desde los archivos CSV
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
df2
```

```python
rename_dict = {
    'Customer': 'customer',
    'ST': 'state',
    'State': 'state',
    'Gender': 'gender',
    'GENDER': 'gender',
    'Education': 'education',
    'Customer Lifetime Value': 'customer_lifetime_value',
    'Income': 'income',
    'Monthly Premium Auto': 'monthly_premium_auto',
    'Number of Open Complaints': 'number_of_open_complaints',
    'Total Claim Amount': 'total_claim_amount',
    'Policy Type': 'policy_type',
    'Vehicle Class': 'vehicle_class'
}

# Estandarizar los nombres de las columnas para df2 utilizando .rename
df2.rename(columns=rename_dict, inplace=True)

# Estandarizar los nombres de las columnas para df3 utilizando .rename
df3.rename(columns=rename_dict, inplace=True)

df2
```

```python
df3
```

```python
columns_df1 = set(df1.columns)
columns_df2 = set(df2.columns)
columns_df3 = set(df3.columns)

print("Columnas en df1:")
print(columns_df1)

print("\nColumnas en df2 despuÃ©s de la estandarizaciÃ³n:")
print(columns_df2)

print("\nColumnas en df3 despuÃ©s de la estandarizaciÃ³n:")
print(columns_df3)
```

```python
df2.state.unique()
```

```python
state_replacements = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington'
}
df2['state'] = df2['state'].replace(state_replacements)
```

```python
# Reemplazar los valores incorrectos en la columna `education` en ambos DataFrames
df2['education'] = df2['education'].replace('Bachelors', 'Bachelor')
df3['education'] = df3['education'].replace('Bachelors', 'Bachelor')
```

```python
# Reemplazar los valores en la columna 'gender' de df2
df2['gender'] = df2['gender'].fillna('U').replace({
    'female': 'F',
    'Femal': 'F',
    'Male': 'M'
})
```

```python
# Estandarizar los valores en la columna 'vehicle_class' en ambos DataFrames
vehicle_class_replacements = { 'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury' }
df2['vehicle_class'] = df2['vehicle_class'].replace(vehicle_class_replacements)
df3['vehicle_class'] = df3['vehicle_class'].replace(vehicle_class_replacements)
```
